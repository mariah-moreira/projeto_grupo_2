```python
# Criar um DataFrame de exemplo (substitua isso pelo seu DataFrame)
data = {
    'Categoria': ['Júnior', 'Pleno', 'Sênior'],
    'Masculino': [20, 30, 15],
    'Feminino': [15, 25, 10]
}
df = pd.DataFrame(data)

# Configuração dos valores para o gráfico
categorias = df['Categoria']
masculino = df['Masculino']
feminino = df['Feminino']

# Definir a largura das barras
largura_barra = 0.35

# Criar o índice para os grupos de barras
indices = range(len(categorias))

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(indices, masculino, width=largura_barra, label='Masculino')
plt.bar([i + largura_barra for i in indices], feminino, width=largura_barra, label='Feminino')

# Configurações do gráfico
plt.xlabel('Categoria')
plt.ylabel('Número de Profissionais')
plt.title('Contagem por Categoria e Gênero')
plt.xticks([i + largura_barra / 2 for i in indices], categorias)
plt.legend()

plt.tight_layout()
plt.show()

```

```python
# Dados de exemplo
anos = [2010, 2011, 2012, 2013, 2014]
receitas = [300, 400, 500, 600, 700]

# Criar o gráfico de área
plt.fill_between(anos, receitas, color='skyblue', alpha=0.4)
plt.plot(anos, receitas, marker='o', color='blue')

# Configurações do gráfico
plt.title('Receitas Anuais')
plt.xlabel('Ano')
plt.ylabel('Receitas ($)')
plt.grid(True)

# Mostrar o gráfico
plt.show()
```
